## Mapa de la ciudad 

Otro objeto de nuestro inters tambin ha sido crear un mapa de la ciudad con las ubicaciones de las empresas, informacin que no devuelve la plataforma de Linkedin. 

A travs del paquete **BeautifulSoup**, hemos buscado la informacin necesaria en *Google* para cada fila de nuestros conjunto de trabajos encontrado. Supuestamente, hemos encontrado unos problemas:

1. La ubicacin puede no encontrarse para todas las empresas con la misma frmula de bsqueda.
2. Si la empresa tiene varias sucursales, se pueden obtener direcciones ms cercanas a la localizacin del dispositivo.

### Busqueda de direccin

```python
import requests
import random
import bs4

address = []
# Para cada empresa de nuestro conjunto de datos
for company in data['Company']:
# En Google search busca la direccion de la empresa,
# mandando acceptacion de Cookies para acceder a la pagina de los resultados
    try:
        url = "https://google.com/search?q="+ company + "direccion sede Roma"
        request_result = requests.get(url, cookies={'CONSENT':'YES+shp.gws-20220315-0-RC1.es+FX+{}'.format(random.randint(100, 999))})
        soup = bs4.BeautifulSoup(request_result.text, "html.parser")
        if hasattr(soup.find( "div" , class_='BNeawe iBp4i AP7Wnd'), 'text') == False:
            url = "https://google.com/search?q="+ company + "indirizzo Roma"
            request_result = requests.get(url, cookies={'CONSENT':'YES+shp.gws-20220315-0-RC1.es+FX+{}'.format(random.randint(100, 999))})
            soup = bs4.BeautifulSoup(request_result.text, "html.parser")
            if hasattr(soup.find( "div" , class_='BNeawe iBp4i AP7Wnd'), 'text') == False:
                url = "https://google.com/search?q="+ company + "Roma headquarters address"
                request_result = requests.get(url, cookies={'CONSENT':'YES+shp.gws-20220315-0-RC1.es+FX+{}'.format(random.randint(100, 999))})
                soup = bs4.BeautifulSoup(request_result.text, "html.parser")
                if hasattr(soup.find( "div" , class_='BNeawe iBp4i AP7Wnd'), 'text') == False:
                    address.append('NA')
                else:
                    via = soup.find( "div" , class_='BNeawe iBp4i AP7Wnd').text
                    address.append(via)
            else:
                via = soup.find( "div" , class_='BNeawe iBp4i AP7Wnd').text
                address.append(via)
        else:
            via = soup.find( "div" , class_='BNeawe iBp4i AP7Wnd').text
            address.append(via)
    except:
        pass
```

### Geolocalizacon de las direccones

```python
import geopy
from geopy.geocoders import Nominatim 
geolocator = Nominatim(user_agent="My Application")

latitude = []
longitude = []
for a in [a for a in address if str(a) != 'NA']:
    location = geolocator.geocode(a)
    print(location)
    latitude.append(location.latitude)
    longitude.append(location.longitude)
    
locations = pd.DataFrame({'latitude':pd.Series(latitude), 'longitude':pd.Series(longitude)})
```

## Busqueda de tweets

Hemos utilizado la funcin `search_tweets()` para descargar las informacines sobre los tweets de nuestro inters especificando el lugar de donde queramos que vinieran. 

Hemos puesto como palabras de busqueda **#Ukrain invasion** y hemos seleccionado la ubicacion de Mosca, poniendo su longitud y latitud y seleccionando todos los tweets geolocalizados dentro de un radio de 800 millas.

```python
# Set up search query
search_query = "#Ukraine invasion -filter:retweets  "

# Collection of TWEETS
# Get tweets from the API
tweets = tw.Cursor(api.search_tweets,
              q=search_query,
              geocode='55.751244,37.618423,800mi',  #coordinates of Moscow, with a range of 1000 miles we are in East Europe
              lang="en",
              result_type='recent').items(500)

# Store the API responses in a list
tweets_copy = []
for tweet in tweets:
    tweets_copy.append(tweet)
    
print("Total Tweets fetched:", len(tweets_copy))

# Creation of the Dataframe
# intialize the dataframe
tweets_df = pd.DataFrame()

# populate the dataframe
for tweet in tweets_copy:
    hashtags = []
    try:
        for hashtag in tweet.entities["hashtags"]:
            hashtags.append(hashtag["text"])
        text = api.get_status(id=tweet.id, tweet_mode='extended').full_text
    except:
        pass
    tweets_df = tweets_df.append(pd.DataFrame({'user_name': tweet.user.name,
                                               'id':tweet.user.id,
                                               'user_location': tweet.user.location,\
                                               'user_description': tweet.user.description,
                                               'user_verified': tweet.user.verified,
                                               'date': tweet.created_at,
                                               'text': text, 
                                               'hashtags': [hashtags if hashtags else None],
                                               'source': tweet.source}))
    tweets_df = tweets_df.reset_index(drop=True)

# convert DateTime to Date
tweets_df['date'] = pd.to_datetime(tweets_df['date']).dt.date

tweets_df.head()
```

## Limpieza de los tweets

Muchos tweets contienen caracteres o smbolos que no son necesarios para nuestra anlisis. Hemos definido la funcin `cleaner()` para limpiar el contenido de los tweets.

Despus hemos guardado el texto 'limpio' en otra columna llamada **text_clean**.

```python
import re
def cleaner(text):
    text = re.sub("@[A-Za-z0-9]+","",text) #Remove @ sign
    text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text) #Remove http links
    text = " ".join(text.split())
    text = text.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    text = text.replace("often", "").replace("sometimes","").replace("said","").replace("think","")
    text = " ".join(w for w in nltk.wordpunct_tokenize(text)
         if w.lower() in words or not w.isalpha())
    return text
    
tweets_df['text_clean'] = tweets_df['text'].apply(cleaner)
```

## Anlisis de polaridad 

Por primera cosa hemos descargado la libreria **NLTK**, la cual contiene varias utilidades que nos permiten manipular y analizar datos lingsticos de manera eficaz.

```python
# Importing Libraries to analyze the polarity of the tweet
import nltk
nltk.download('vader_lexicon')
nltk.download('words')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

words = set(nltk.corpus.words.words())
```

```python
# Let's perform the Sentiment Analysis
# We create our own dictionary with words and their own polarity(-1 for negatives and 1 for positives)
word_dict = {'blood':-1,'war':-1,'Russian invasion':-1,'teamPutin':-1,
            'bombs':-1,'call to arms':-1,'support Russia':-1,'support Putin':-1,'Nuclear abolished':1,'Nuclear alert':-1,'NATO supports Ukraine':1,
            'support Ukraine':1,'support Kyev':1,'support Volodymyr Zelensky':1,'support Zelensky':1,'team Zelensky':1,'Zelensky resists':1,'Ukraine resists':1 }

sid.lexicon.update(word_dict)

list1 = []
for i in tweets_df['text_clean']:
    list1.append((sid.polarity_scores(str(i)))['compound'])
```

```python
# Coverting the compound scores into categories:"negative", "neutral" or "positive"
# We add the new column 'sentiment' and we fill it with the previous list
tweets_df['sentiment'] = pd.Series(list1)

def sentiment_category(sentiment):
    label = ''
    if(sentiment>0):
        label = 'positive'
    elif(sentiment == 0):
        label = 'neutral'
    else:
        label = 'negative'
    return(label)

tweets_df['sentiment_category'] = tweets_df['sentiment'].apply(sentiment_category)
```
